```python
# Importing Libraries
import re
import pandas as pd
```

```python
# Loading Exported whatsapp chat file
f = open('/content/WhatsApp Chat Kindergarden.txt', 'r', encoding = 'utf-8')
```

```python
# Reading the text file
data = f.read()
```

```python
# Showcasing Data
data
```

```python
print(type(data))
```

```python
# Splitting data based on '\n'
df = data.split('\n')
df
```

```python
# The below pattern variable is a regular expression that defines a specific date and time format.
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
# 're.split(pattern, data)', it splits the 'data' string (which contains the entire WhatsApp chat) at every occurrence of the date and time pattern.
# By using '[1:]' after the split, you are slicing the list to exclude the first element (the empty string) and keep only the actual messages.
messages = re.split(pattern, data)[1:]
messages
```

```python
# dates will find all the excluded first element from whatsapp chat.
dates = re.findall(pattern, data)
dates
```

```python
df = pd.DataFrame({'user_message': messages, 'message_date': dates})
# convert message_date type
# Use a regular expression to remove the trailing hyphen and any surrounding whitespace
df['message_date'] = df['message_date'].str.replace(r'\s*-\s*', '', regex=True)
# Fix the format string to match month before day
df['message_date'] = pd.to_datetime(df['message_date'], format = 'mixed')
# Rename message_date column name to date column
df.rename(columns = {'message_date': 'date'}, inplace = True)
df.head()
```

```python
df.shape
```

```python
# Separate username and messages
users = []
messages = []

for message in df['user_message']:
  entry = re.split('([\w\W]+?):\s', message)
  if entry[1:]:
    users.append(entry[1])
    messages.append(entry[2])
  else:
    users.append('group_notification')
    messages.append(entry[0])


df['user'] = users
df['message'] = messages
df.drop(columns = ['user_message'], inplace = True)

df.head()
```

```python
# Here to exclude all the stop words from whatsapp chat messages we are using 'stop_hinglist.txt' file.
f = open('/content/stop_hinglish.txt', 'r')
stop_words = f.read()
print(stop_words)
```

```python
# words variable will contain only words from whatsapp chat messages which are not present in stop words file.
words = []

for message in temp['message']:
  for word in message.lower().split():
    if word not in stop_words:
      words.append(word)
```

```python
# period variable have hour period with next hour
period = []

for hour in df[['day_name', 'hour']]['hour']:
  if hour == 23:
    period.append(str(hour) + "-" + str('00'))
  elif hour == 0:
    period.append(str('00') + "-" + str(hour + 1))
  else:
    period.append(str(hour) + "-" + str(hour + 1))
```

```python
period
```

```python
# Adding period column to dataset.
df['period'] = period
```

```python
df.head()
```

```python
# creating pivot table day-name as index, period as column, and value as count of messages.
df.pivot_table(index = 'day_name', columns = 'period', values = 'message', aggfunc = 'count').fillna(0)
```
